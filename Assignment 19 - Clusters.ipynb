```python
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
penguin = sns.load_dataset("penguins")
penguin.head()
```

## Pairplot

```python
sns.pairplot(penguin)
```

#### Do you visually notice any clusters? How many clusters do you think there are?
##### Ja. Ik denk dat er twee clusters zijn.

## Filling & Cluster Model

```python
value = {'bill_length_mm': penguin['bill_length_mm'].mean(), 'bill_depth_mm': penguin['bill_depth_mm'].mean(), 
         'flipper_length_mm': penguin['flipper_length_mm'].mean(), 'body_mass_g': penguin['body_mass_g'].mean(), 
         'sex': 'Unkown'}
penguin = penguin.fillna(value = value)
```

```python
features = ['body_mass_g','flipper_length_mm','bill_depth_mm', 'bill_length_mm']
km = KMeans(n_clusters=2, random_state=42).fit(penguin[features])
```

```python
penguin['cluster'] = km.predict(penguin[features])
```

```python
penguin.head()
```

```python
penguin.cluster.value_counts()
```

```python
sns.pairplot(penguin, hue="cluster")
```

## Silhouette Coefficient

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

```python
# You can change the features and n_clusters in the k-means above and check the impact on the Silhouette Coefficient
metrics.silhouette_score(penguin[features], km.labels_, metric='euclidean')
```

##### Clusters = 2 geeft 0.627, Clusters = 3 geeft 0.579, Clusters = 4 geeft 0,0554

## Contingency Table

```python
sns.pairplot(penguin, hue="species")
```

```python
contingency_table = penguin.groupby(['species','cluster']).size().unstack('cluster', fill_value=0)
contingency_table
```
