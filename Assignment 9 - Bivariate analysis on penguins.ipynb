```python
import pandas as pd
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.corr()
```

```python
penguinsCorr = penguins.corr()
penguinsCorr.style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```

```python
penguinsCorr.plot(figsize=(15,8))
```
