```python
import pandas as pd
lifeExpectancy = pd.read_csv('Datasets/lifeExpectancyAtBirth.csv', sep=',')
```

```python
lifeExpectancy.head()
```

```python
periodAndDim1Filter = (lifeExpectancy['Period'] == 2019) & (lifeExpectancy['Dim1'] == "Both sexes")
```

```python
result = lifeExpectancy[periodAndDim1Filter]
```

```python
result.head()
```
