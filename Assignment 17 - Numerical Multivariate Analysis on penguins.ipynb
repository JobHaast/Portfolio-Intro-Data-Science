```python
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
penguin = sns.load_dataset("penguins")
penguin.head()
```

## Test/train split

```python
value = {'bill_length_mm': penguin['bill_length_mm'].mean(), 'bill_depth_mm': penguin['bill_depth_mm'].mean(), 
         'flipper_length_mm': penguin['flipper_length_mm'].mean(), 'body_mass_g': penguin['body_mass_g'].mean(), 
         'sex': 'Unkown'}
penguin = penguin.fillna(value = value)
```

```python
penguin_train, penguin_test = train_test_split(penguin, test_size=0.3, stratify=penguin['species'], random_state=42)
print(penguin_train.shape, penguin_test.shape)
```

```python
features= ['bill_length_mm']
dt_regression = DecisionTreeRegressor(max_depth = 3) # Increase max_depth to see effect in the plot
dt_regression.fit(penguin_train[features], penguin_train['body_mass_g'])
```

```python
from sklearn import tree
import graphviz

def plot_tree_regression(model, features):
    # Generate plot data
    dot_data = tree.export_graphviz(model, out_file=None, 
                          feature_names=features,  
                          filled=True, rounded=True,  
                          special_characters=True)  

    # Turn into graph using graphviz
    graph = graphviz.Source(dot_data)  

    # Write out a pdf
    graph.render("decision_tree_numerical_penguins")

    # Display in the notebook
    return graph 
```

```python
plot_tree_regression(dt_regression, features)
```

```python
def calculate_rmse(predictions, actuals):
    if(len(predictions) != len(actuals)):
        raise Exception("The amount of predictions did not equal the amount of actuals")
    
    return (((predictions - actuals) ** 2).sum() / len(actuals)) ** (1/2)

## The same function but using a for-loop instead of a vectorized operation. 
# def calculate_rmse(predictions, actuals):
#    if(len(predictions) != len(actuals)):
#        raise Exception("The amount of predictions did not equal the amount of actuals")
#    
#    diffSquared = 0
#    
#    for prediction_i, actual_i in zip(predictions, actuals):
#        diffSquared += (prediction_i - actual_i)**2
#        
#    return (diffSquared/len(actuals))**(1/2)
```

```python
predictionsOnTrainset = dt_regression.predict(penguin_train[features])
predictionsOnTestset = dt_regression.predict(penguin_test[features])

rmseTrain = calculate_rmse(predictionsOnTrainset, penguin_train.body_mass_g)
rmseTest = calculate_rmse(predictionsOnTestset, penguin_test.body_mass_g)

print("RMSE on training set " + str(rmseTrain))
print("RMSE on test set " + str(rmseTest))
```
